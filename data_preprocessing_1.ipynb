# import

```python
import pandas as pd
import numpy as np
```

# manual binning

```python
values_nama = ['joko','jono','giorno','joseph','jotaro','jonathan']
values_status_r = ['kontrak','milik','kontrak','kontrak','milik','kontrak']
values_gaji_ot = [2100000,4000000,1400000,700000,650000,450000]
values_keterangan = ['','','','','','']

df = pd.DataFrame({
    'nama' : values_nama,
    'status rumah' : values_status_r,
    'gaji orang tua' : values_gaji_ot,
    'keterangan' : values_keterangan
    })

df.head(6)
```

```python
batas_bin = [0,1400000,4000000]
kategori = ['kecil','besar']

df['gaji_binned_1'] = pd.cut(df['gaji orang tua'], bins=batas_bin, labels=kategori)
```

```python
print(df)
```

# linspace

```python
bins = np.linspace(
    min(df['gaji orang tua']), 
    max(df['gaji orang tua']), 3)
print(bins)
```

```python
kategori = ['kecil','besar']
```

```python
df['gaji_binned'] = pd.cut(df['gaji orang tua'], bins=bins, labels=kategori, include_lowest=True)
print(df)
```

# quantile

```python
df['gaji_binned3'] = pd.cut(df['gaji orang tua'], 2)
print(df)
```

```python
kategori = ['kecil', 'besar']
df['gaji_binned3'] = pd.qcut(df['gaji orang tua'], q=2, labels=kategori)
print(df)
```

# feature engineering

## imputation

```python
values_nama = ['joko','jono','giorno','joseph','jotaro','jonathan']
values_status_r = ['kontrak','milik','kontrak','kontrak','milik','kontrak']
values_gaji_ot = [None,4000000,1400000,700000,None,450000]
values_keterangan = ['diterima','ditolak','ditolak','diterima','diterima','diterima']

df = pd.DataFrame({
    'nama' : values_nama,
    'status rumah' : values_status_r,
    'gaji orang tua' : values_gaji_ot,
    'keterangan' : values_keterangan
    })

df.head(6)
```

```python
df = df.fillna(0)
```

```python
df = df.fillna(df.median())
df.head()
```

```python
df['gaji orang tua'].fillna(df['gaji orang tua'].value_counts().idxmax(), inplace=True)
```

# feature selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('cardata-germany-dataset.csv')
df.head()
```

```python
_shape = df.shape
_dtypes = df.dtypes
print(f'shape of df : {_shape}\ndtypes of df : {_dtypes}')
```

```python
df = df.drop('offerType', axis=1)
df.head()
```

```python
make_model = pd.DataFrame({
    'make' : df['make'].astype(str), 
    'model' :df['model'].astype(str)
})
make_model.head()
```

```python
df = df.drop(['make','model'], axis=1)
df.head()
```

```python
df = df.dropna()
df.head
```

```python
fuel_unique = pd.DataFrame(df['fuel'].sort_values().unique())
fuel_unique = fuel_unique.replace({r'[^\w]':''}, regex=True)
print(fuel_unique)

gear_unique = pd.DataFrame(df['gear'].sort_values().unique())
print(gear_unique)
```

```python
fu_1hot = OneHotEncoder(sparse=False)
fu_1hot = fu_1hot.fit_transform(fuel_unique)

fu_1hot = pd.DataFrame(fu_1hot)
header = fuel_unique
fu_1hot.columns = header
fu_1hot.head()
```

```python
gu_1hot = OneHotEncoder(sparse=False)
gu_1hot = gu_1hot.fit_transform(gear_unique)

gu_1hot = pd.DataFrame(gu_1hot)
header = gear_unique
gu_1hot.columns = header
gu_1hot.head()
```

```python
data_join = df.join([fu_1hot,gu_1hot])
data_join.head()
```
